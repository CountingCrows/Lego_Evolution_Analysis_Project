### Importing and Reading files from Github


```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CountingCrows/legoanalysisproject/main/dataset/lego_sets.csv')
df.head()
```

```python
theme = pd.read_csv('https://raw.githubusercontent.com/CountingCrows/legoanalysisproject/main/dataset/parent_themes.csv')
theme.head()
```

#### Task 1: What percentage of all liscensed sets ever released were Star Wars Themed?

```python
## merging syntaxs using pandas
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged.drop(columns='name_y', inplace=True)
merged.head()
```

```python
## dealing with missing values
merged[merged['set_num'].isnull()]
```

```python
## dealing with missing values
merged[merged['set_num'].isnull()].shape
```

```python
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
```

```python
## check again for missing values in set_num
licensed[licensed['set_num'].isnull()].shape
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head(50)
```

```python
star_wars.shape
```

```python
the_force = int(star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
```

#### The percentage of all liscensed sets ever released were Star Wars Themed is 51%

### Task 2: In which year was Star Wars not the most popular licensed theme?

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted.groupby(['year','parent_theme']).head(10)
```

```python
licensed_sorted['count'] = 1

summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()

summed_df.head()
```

```python
## get only max row with the same 'year'
max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])

max_df.sort_values('year', inplace=True)

max_df
```

```python
new_era = 2017
```

### The year when Star Wars not the most popular licensed theme is in 2017, it's place was taken by Super Heroes theme

### Task 3: How many unique sets were released each year (1995-2017)?

```python
## break down number of sets by year
clean_df = merged[~merged['set_num'].isnull()]

clean_df['count'] = 1

sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year', 'count']]

sets_per_year.head(50)
```

```python
for index, row in sets_per_year.iterrows():
    print(row['year'], row['count'])
```
